**String Cleaning**

```python
import pandas as pd

data = {
    "Name": ["Chandu", "Suresh", "kishore", "shiva", "vinay", "mural", "karthik", "deeven", "satish"],
    "Age": [45, 24, 34, 35, 23, 25, 24, 25, 25],
    "City": ["kkd", "kkd", "vzg", "kkd", "kkd", "kkd", "anakapalli", "kkd", "kkd"]
}

df = pd.DataFrame(data)
print(df)
```

```python
df['City']=df['City'].str.upper()
df
```

```python
df['City']=df['City'].str.lower()
df
```

```python
df['Name'].str.lower()
```

```python
df['City']=df['City'].str.replace("Anakapalli","RJY")
df
```

**Concatenate DataFrames**

```python
india=pd.DataFrame({
    "city": ["Mumbai","Agra","Banglore"],
    "temp": [34,23,45]
})
india
```

```python
us=pd.DataFrame({
    "city": ["Chicago", "Newyork"," Ireland"],
    "temp": [45,23,67]
})
us
```

```python
df=pd.concat([india,us])
df
```

```python
df=pd.concat([india,us],ignore_index=True) # it gives line by line ignore index true means it follow ordered index
df
```

```python
df= pd.concat([india,us],axis=1) # it gives side by side tables
df
```

**Merge dataframes**

```python
temp=pd.DataFrame({
    "city":["hyd","vzg","N"],
    "tempt":[34,45,67]
})
temp
```

```python
t=pd.DataFrame({
    "city":["hyd","vzg","vijy"],
    "rainfall":[36,55,77]
})
t
```

```python
df=pd.merge(temp,t,on="city")  #it merges two datas and gives the common columns
df
```

```python
df=pd.merge(temp,t,on="city",how="outer")  #it gives all with null values
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df["rainfall"]=df["rainfall"].fillna(df.rainfall.min())
df
```

```python
df["tempt"]=df["tempt"].fillna(df.tempt.max())
df
```
