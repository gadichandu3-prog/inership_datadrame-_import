```python
import pandas as pd

# Sample DataFrames
df1 = pd.DataFrame({
    "Name": ["Kumar", "Ashok"],
    "Age": [20, 17]
})

df2 = pd.DataFrame({
    "Name": ["Mural", "Krishna"],
    "Age": [18, 19]
})

# Concatenate vertically (stack rows)
df_concat = pd.concat([df1, df2])

print(df_concat)
```

MinMax

```python
import pandas as pd

# Sample DataFrame
df = pd.DataFrame({
    "Name": ["Kumar", "Ashok", "Mural", "Krishna", "Swamy"],
    "Marks": [20, 40, 25, 60, 45],
    "Age": [20, 17, 18, 19, 20]
})

# Minimum marks
min_marks = df["Marks"].min()
print("Minimum Marks:", min_marks)

# Maximum marks
max_marks = df["Marks"].max()
print("Maximum Marks:", max_marks)
```

```python
import pandas as pd

# Sample DataFrames
df1 = pd.DataFrame({
    "ID": [1, 2, 3],
    "Name": ["Kumar", "Ashok", "Mural"]
})

df2 = pd.DataFrame({
    "ID": [1, 2, 4],
    "Marks": [50, 60, 70]
})

# Merge on ID (inner join by default)
df_merged = pd.merge(df1, df2, on="ID")
print(df_merged)
```

```python
import pandas as pd

# Sample DataFrame
df = pd.DataFrame({
    "Name": ["Kumar", "Ashok", "Mural", "Kumar"],
    "Gender": ["Male", "Male", "Male", "Male"],
    "Dept": ["A", "B", "A", "C"]
})

# Encode Dept column (Label Encoding)
df["Dept_encoded"] = df["Dept"].map({"A": 1, "B": 2, "C": 3})
print(df)
```

```python
import pandas as pd

# Sample DataFrame
df = pd.DataFrame({
    "Name": ["Kumar", "Ashok", "Mural", "Swamy"],
    "Experience_Level": ["Junior", "Senior", "Mid", "Junior"]
})

# Define ordinal mapping
level_mapping = {"Junior": 1, "Mid": 2, "Senior": 3}

# Apply mapping
df["Experience_Level_encoded"] = df["Experience_Level"].map(level_mapping)

print(df)
```

```python
import pandas as pd

# Sample DataFrame
df = pd.DataFrame({
    "Dept": ["A", "B", "A", "C", "B", "A", "C"]
})

# Calculate frequency of each category
freq_encoding = df["Dept"].value_counts() / len(df)

# Map frequency to the column
df["Dept_freq_encoded"] = df["Dept"].map(freq_encoding)

print(df)
```
